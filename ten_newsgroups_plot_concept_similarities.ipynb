# Plotting concept similarities

The CW module aims to separate the concepts in the latent space. In order to visualize the separability of the concepts' latent representations, the intra-concept similarity and the inter-concept similarity metrics are used.

The latent representation of each graph concept from the output of the CW module is first extracted. The representation for instance $j$ of concept $i$ is denoted as $\mathbf{x}_{ij}$. The intra-concept similarity for concept $i$, denoted as $d_{ii}$ is then defined to be

\begin{equation} 
    d_{ii} = \frac{1}{n^{2}} \left( \sum_{j=1}^{n} \sum_{k=1}^{n} \frac{\mathbf{x}_{ij} \cdot \mathbf{x}_{ik}}{\lVert \mathbf{x}_{ij} \rVert_2 \lVert \mathbf{x}_{ik} \rVert_2} \right)
\end{equation}

where $n$ is the total number of instances of concept $i$.

Inter-concept similarity between concept $p$ and $q$ is similarly defined as

\begin{equation} 
    d_{pq} = \frac{1}{nm} \left( \sum_{j=1}^{n} \sum_{k=1}^{m} \frac{\mathbf{x}_{pj} \cdot \mathbf{x}_{qk}}{\lVert \mathbf{x}_{pj} \rVert_2 \lVert \mathbf{x}_{qk} \rVert_2} \right)
\end{equation}

where $n$ and $m$ are the number of instances of concepts $p$ and $q$, respectively.

Intra-concept similarity is the average pair-wise cosine similarity between instances of the same concept, and inter-concept similarity is the average pairwise cosine similarity between instances of two different concepts.

Heat maps can then be defined, where the value in a cell at row $i$ column $j$ is computed as

\begin{equation} 
    Q_{ij} = \frac{d_{ij}}{\sqrt{d_{ii}d_{jj}}}
\end{equation}
